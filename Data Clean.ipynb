```python
import pandas as pd
```

```python
meta_df = pd.read_csv("case study/GDP/Metadata_Country.csv")
GDP_df = pd.read_csv("case study/GDP/GDP.csv")
Pop_df = pd.read_csv("case study/Population/Population.csv")
```

```python
Pop_df.to_csv("Data/Pop.csv")
```

```python
GDP_df = GDP_df.drop(['Indicator Name','Indicator Code','Unnamed: 66'], axis =1)
Pop_df= Pop_df.drop(['Indicator Name','Indicator Code','Unnamed: 66'], axis =1)
```

```python
GDP_df = pd.merge(GDP_df, meta_df)
Pop_df = pd.merge(Pop_df, meta_df)
```

```python
Pop_df = Pop_df.drop(['Unnamed: 5','TableName','SpecialNotes'],axis = 1)
GDP_df = GDP_df.drop(['Unnamed: 5','TableName','SpecialNotes'],axis = 1)
```

```python
Pop_df.to_csv("Data/Pop.csv")
GDP_df.to_csv("Data/GDP.csv")
```
